```python
import anthropic
import os

api_key = os.environ.get("ANTHROPIC_API_KEY", "your-key-here")
client = anthropic.Anthropic(api_key=api_key)

message = client.messages.create(
    model="claude-opus-4-6",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Hello! Are you working?"}
    ]
)

print(message.content[0].text)
```

```python
def symptom_checker(symptoms):
    message = client.messages.create(
        model="claude-opus-4-6",
        max_tokens=1024,
        system="You are a helpful medical assistant. When given symptoms, provide possible conditions they could indicate, when to see a doctor, and simple home care tips. Always remind the user you are an AI and not a substitute for professional medical advice.",
        messages=[
            {"role": "user", "content": f"I have the following symptoms: {symptoms}. What could this indicate?"}
        ]
    )
    return message.content[0].text

# Test it
symptoms = "fever, sore throat, and body aches"
result = symptom_checker(symptoms)
print(result)
```

```python
def chat_with_doctor():
    conversation_history = []

    print("Welcome to AI Health Assistant!")
    print("Type 'quit' to exit\n")
    print("-" * 50)

    while True:
        user_input = input("You: ")

        if user_input.lower() == 'quit':
            print("Take care and stay healthy!")
            break

        conversation_history.append({
            "role": "user",
            "content": user_input
        })

        response = client.messages.create(
            model="claude-opus-4-6",
            max_tokens=1024,
            system="You are a friendly medical assistant. Ask for the patient's age and relevant follow up questions before giving a diagnosis. Give detailed explanations and what tests are required. Always remind you are an AI and not a substitute for professional medical advice.",
            messages=conversation_history
        )

        assistant_message = response.content[0].text

        conversation_history.append({
            "role": "assistant",
            "content": assistant_message
        })

        print(f"\nDoctor AI: {assistant_message}\n")
        print("-" * 50)

chat_with_doctor()
```

```python
import uuid
import json
from datetime import datetime

def chat_with_doctor():
    conversation_history = []

    patient_id = str(uuid.uuid4())[:6].upper()
    print(f"Your Patient ID is: {patient_id} — please save this for future visits!")

    print("Welcome to AI Health Assistant!")
    print("Type 'quit' to exit\n")
    print("-" * 50)

    patient_name = input("enter your name: ")


    while True:
        user_input = input("You: ")

        if user_input.lower() == 'quit':
            # Save conversation to file
            filename = f"conversation_{patient_name}_{patient_id}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
            with open(filename, 'w') as f:
                json.dump(conversation_history, f, indent=2)
            print(f"\nConversation saved to {filename}")
            print("Take care and stay healthy!")
            break

        conversation_history.append({
            "role": "user",
            "content": user_input
        })

        response = client.messages.create(
            model="claude-opus-4-6",
            max_tokens=1024,
            system="You are a friendly medical assistant. Ask for the patient's age and relevant follow up questions before giving a diagnosis. Give detailed explanations and what tests are required. Always remind you are an AI and not a substitute for professional medical advice.",
            messages=conversation_history
        )

        assistant_message = response.content[0].text

        conversation_history.append({
            "role": "assistant",
            "content": assistant_message
        })

        print(f"\nDoctor AI: {assistant_message}\n")
        print("-" * 50)

chat_with_doctor()
```
